
# Intelligent Car Price Prediction System

## Project Overview
This project focuses on building a machine learning system capable of estimating car prices based on key vehicle attributes such as engine size, horsepower, fuel type, and other structural features.

Car pricing in many markets is often inconsistent and influenced by subjective judgment. This project demonstrates how data-driven models can provide more reliable and transparent price estimates.



## Problem Statement
Determining the fair price of a vehicle is challenging due to variations in features and market conditions. This project aims to solve this by predicting prices using measurable data, reducing reliance on guesswork.



## Approach
- Data preprocessing and cleaning
- Feature engineering (brand extraction)
- Model training and comparison
- Evaluation using R², MAE, RMSE



## Models Implemented
- Linear Regression
- Decision Tree
- Random Forest (Best Performing)
- XGBoost
- Neural Network



## Key Results
Random Forest achieved the best performance:
- R² Score: 0.9570
- Lowest MAE and RMSE

This confirms that ensemble methods are highly effective for modeling non-linear relationships in pricing data.



## Key Insights
- Engine size, horsepower, and curb weight are major price drivers
- Ensemble models outperform simple linear models
- Feature engineering significantly improves results



## Real-World Relevance
This system can be used to improve pricing transparency in vehicle markets, especially in regions where pricing is inconsistent and lacks standardization.
